```python
import pandas as pd

#fname = "/Users/simon.hughes/Google Drive/PhD/Data/CoralBleaching/Results/predictions.txt"
fname = "/Users/simon.hughes/Google Drive/PhD/Data/CoralBleaching/Results/predictions-causal_relations.txt"

data = pd.read_csv(fname, sep="|")
```

```python
data.head()
```

```python
data["Concept Codes"] = data["Concept Codes"].apply(str)
data["Predictions"] = data["Predictions"].apply(str)
```

```python
code = "Causer:5->Result:50"

#with_ys = data["Concept Codes"].str.contains(code)
#without = with_ys[not data["Concept Codes"].str.contains(code)]
df = data[data["Concept Codes"].str.contains(code) & ~data["Predictions"].str.contains(code)]
def filter_causal(s):
    splt = s.split(",")
    return "\t".join(filter(lambda st: "->" in st and "_" not in st, splt)).\
        replace("Causer:","C:").replace("Result:","R:")

print code, len(df)
for i in range(len(df)):
    row = df.iloc[i]
    ix = df.iloc[[i]].index.tolist()[0] + 2
    #print dir(row)
    print "Line:  ", str(ix), "Essay:", row["Essay"]
    print "Sent:  ", row["Processed Sentence"]
    print "Codes: ", row["Concept Codes"]
    print "Preds: ", row["Predictions"]
    print "CCodes:", filter_causal(row["Concept Codes"])
    print "CPreds:", filter_causal(row["Predictions"])
    print
```
